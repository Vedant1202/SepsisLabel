### Deep net Classification

```python
dfme_deep = dfme.drop(['Unit1', 'Unit2', 'HospAdmTime'], axis=1)
```

```python
dfme_deep.head()
```

```python
import tensorflow as tf
```

```python
dfme_deep.shape
```

```python
dfme_deep.columns
```

```python
hrCol = tf.feature_column.numeric_column('HR')
o2satCol = tf.feature_column.numeric_column('O2Sat')
sbpCol = tf.feature_column.numeric_column('SBP')
mapCol = tf.feature_column.numeric_column('MAP')
dbpCol = tf.feature_column.numeric_column('DBP')
respCol = tf.feature_column.numeric_column('Resp')
ageCol = tf.feature_column.numeric_column('Age')
genderCol = tf.feature_column.numeric_column('Gender')
iculosCol = tf.feature_column.numeric_column('ICULOS')
```

```python
feat_cols = [hrCol, o2satCol, sbpCol, mapCol, dbpCol, respCol, ageCol, genderCol,iculosCol]
```

```python
X_train_dnn, X_test_dnn, y_train_dnn, y_test_dnn = train_test_split(dfme_deep.drop(['SepsisLabel', 'patientID'], axis=1), dfme_deep['SepsisLabel'], test_size=0.2, random_state=42)
```

### LinearClassifier

```python
input_fn = tf.estimator.inputs.pandas_input_fn(x=X_train_dnn, y=y_train_dnn,
                                              batch_size=10, num_epochs=10000,
                                              shuffle=True)
```

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols, n_classes=2)
```

```python
model.train(input_fn=input_fn, steps=10000)
```

```python
eval_input_fn = tf.estimator.inputs.pandas_input_fn(x=X_test_dnn, y=y_test_dnn, 
                                                   batch_size=10,
                                                   num_epochs=1,
                                                   shuffle=False)
```

```python
resultsLinClass = model.evaluate(eval_input_fn)
```

```python
resultsLinClass
```

```python
pred_input_fn = tf.estimator.inputs.pandas_input_fn(x=X_test_dnn,
                                                   batch_size=10,
                                                   num_epochs=1,
                                                   shuffle=False)
```

```python
predsLinClass = model.predict(pred_input_fn)
```

```python
predLC = list(predsLinClass)
```

```python
predLC
```

```python
predsLCFinal = [int(str(i['classes'][0])[2:-1]) for i in predLC]
```

```python
pd.Series(predsLCFinal).unique()
```

```python
print(classification_report(y_test_dnn, predsLCFinal))
```

### DNNClassifier

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[10, 20, 20, 20, 10], 
                                       feature_columns=feat_cols, n_classes=2)
```

```python
dnn_input_fn = tf.estimator.inputs.pandas_input_fn(X_train_dnn, y_train_dnn, batch_size=5,
                                                  num_epochs=10000, shuffle=True)
```

```python
dnn_model.train(dnn_input_fn, steps=10000)
```

```python
dnn_eval_input_fn = tf.estimator.inputs.pandas_input_fn(X_test_dnn, y_test_dnn,
                                                       batch_size=5, 
                                                       num_epochs=1,
                                                       shuffle=False)
```

```python
dnn_model.evaluate(eval_input_fn)
```

```python
dnn_pred_input_fn = tf.estimator.inputs.pandas_input_fn(x=X_test_dnn,
                                                   batch_size=5,
                                                   num_epochs=1,
                                                   shuffle=False)
```

```python
predsDNN = dnn_model.predict(dnn_pred_input_fn)
```

```python
predDN = list(predsDNN)
```

```python
predDN
```

```python
predsDNNFinal = [int(str(i['classes'][0])[2:-1]) for i in predDN]
```

```python
pd.Series(predsDNNFinal).unique()
```

```python
print(classification_report(y_test_dnn, predsDNNFinal))
```

```python
z=0
o=0
t=0
for i in range(len(predsDNNFinal)):
    if predsDNNFinal[i] == y_test_dnn.tolist()[i]:
        t += 1
        if predsDNNFinal[i] == 1:
            o += 1
        else: 
            z += 1
```
